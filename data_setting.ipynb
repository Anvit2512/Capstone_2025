```python
import glob

csv_files = glob.glob("data/*.csv")
print("Number of CSV files:", len(csv_files))
```

```python
import os

DATA_DIR = "data"

# Get all CSV files (sorted for consistency)
csv_files = sorted([f for f in os.listdir(DATA_DIR) if f.lower().endswith(".csv")])

if not csv_files:
    print("No CSV files found in data/")
    exit()

print(f"Found {len(csv_files)} CSV files\n")

print(f"{'OLD NAME':40s} {'NEW NAME':10s} {'SIZE (KB)':>10s}")
print("-" * 65)

for idx, old_name in enumerate(csv_files, start=1):
    old_path = os.path.join(DATA_DIR, old_name)

    # File size in KB
    size_kb = os.path.getsize(old_path) / 1024.0

    new_name = f"{idx}.csv"
    new_path = os.path.join(DATA_DIR, new_name)

    # Rename
    os.rename(old_path, new_path)

    print(f"{old_name:40s} {new_name:10s} {size_kb:10.2f}")

print("\n✅ Renaming complete.")
```
